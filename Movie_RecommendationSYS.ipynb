```python
import ast
```

```python
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies_df['genre'] =movies_df['genres'].apply(convert)
movies_df['genre'] =movies_df['keywords'].apply(convert)
movies_df.head()
```

```python
def convert3(obj):
  L=[]
  counter = 0
  for i in ast.literal_eval(obj):
    if counter !=3:
      L.append(i['name'])
      counter +=1
    else:
      break
    return L
```

```python
movies_df['cast']=movies_df['cast'].apply(convert3)
```

```python
movies_df.head()
```

```python
def fetch_director(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      L.append(i['name'])
  return L
```

```python
movies_df['crew']=movies_df['crew'].apply(fetch_director)
```

```python
movies_df['overview'] = movies_df['overview'].astype(str)
movies_df['overview'] = movies_df['overview'].apply(lambda x:x.split())
```

```python
movies_df['genres'] = movies_df['genres'].apply(lambda x:x.split())
```

```python
movies_df['keywords'] = movies_df['keywords'].apply(lambda x:x.split())
```

```python
def clean_cast(x):
    if x is None:
        return []
    else:
        return [i.replace(" ","") for i in x]

movies_df['cast'] = movies_df['cast'].apply(clean_cast)
```

```python
movies_df['crew'] = movies_df['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies_df.head()
```

```python
movies_df['tags']= movies_df['overview']+movies_df['genres']+movies_df['keywords']+movies_df['cast']+movies_df['crew']
```

nltk= natural language tool kit is a platform used build python program that work with human language

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

**creating recommend function to check the movie recommendation system**

```python
def recommend(movie):
  movie_index =new_df[new_df['title']== movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse= True, key = lambda x:x[1])[1:6]
  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Iron Man')
```
